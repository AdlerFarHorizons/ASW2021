```python
#file1 = open("JesusGarcia _07192021_15_00-07202021_10_45_cleaned.CSV","r")
#file1 = open("ApoorvaSomani_07192021_1600-07202021_0830_CLEANED.CSV","r")

file1 = open("DYuen RGB+SQM Data 07-20-2021 01_30-08_00.CSV.CSV","r")
file1.readline()    # read the headers

i=0
Time=[]
SQM=[]
Red=[]
Green=[]
Blue=[]
CT=[]
Lux=[]
r=[]
g=[]
b=[]
c=[]
for line in file1:
    ld=line.split(",")
    if len(ld)==11: 
        Time.append(int(ld[0]))
        SQM.append(float(ld[1]))
        Red.append(float(ld[2]))
        Green.append(float(ld[3]))
        Blue.append(float(ld[4]))
        CT.append(int(ld[5]))
        Lux.append(int(ld[6]))
        r.append(int(ld[7]))
        g.append(int(ld[8]))
        b.append(int(ld[9]))
        c.append(int(ld[10]))

#    print(i,ld)
#    if i>100:
#        break
#    i=i+1

Time=np.array(Time)
SQM=np.array(SQM)
Red=np.array(Red)
Green=np.array(Green)
Blue=np.array(Blue)
CT=np.array(CT)
Lux=np.array(Lux)
r=np.array(r)
g=np.array(g)
b=np.array(b)
c=np.array(c)

file1.close()
```
